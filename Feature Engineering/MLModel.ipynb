```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Housing_Final.csv')
```

```python
df.head()
```

```python
df.drop(['Unnamed: 0', 'Unnamed: 0.1'], axis = 1)
```

```python
df.info()
```

```python
X = df.drop('SalePrice', axis = 1)
```

```python
y = df['SalePrice']
```

```python
from sklearn.model_selection import train_test_split
```

```python
 X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_X_train = scaler.fit_transform(X_train)
```

```python
scaled_X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import ElasticNet
```

```python
base_elastic_model = ElasticNet()
```

```python
param_grid = {'alpha':[0.1,1,5,10,100], 'l1_ratio' : [.1, .7, .99, 1]}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_model = GridSearchCV(base_elastic_model, param_grid=param_grid, scoring = 'neg_mean_squared_error', cv=5, verbose = 1)
```

```python
grid_model.fit(X_train,y_train)
```

```python
# displaying the combination of best parameters for model
```

```python
grid_model.best_params_
```

```python
# Evaluating Model Prediction
```

```python
y_pred = grid_model.predict(scaled_X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
MAE = mean_absolute_error(y_test, y_pred)
```

```python
MAE
```

```python
RMSE = np.sqrt(mean_squared_error(y_test, y_pred))
```
